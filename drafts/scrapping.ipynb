```python
from langchain_community.document_loaders import RecursiveUrlLoader
import bs4
```

```python
def extract_handbook_text(url="https://syloper.com/"):
    """
    Recursively extract text from the Python Data Science Handbook website.
    Uses LangChain's RecursiveUrlLoader with a BeautifulSoup extractor.
    """
    # Using RecursiveUrlLoader to crawl the handbook
    loader = RecursiveUrlLoader(
        url=url, 
        max_depth=4, # To get deeper into chapters
        extractor=lambda x: bs4.BeautifulSoup(x, "html.parser").get_text(separator='\n', strip=True)
    )
    docs = loader.load()
    return docs

# Sample usage:
handbook_docs = extract_handbook_text()
print(f"Successfully extracted {len(handbook_docs)} pages from the handbook.")
```

```python
with open("web_info.txt", "w") as f:
    for item in handbook_docs:
        f.write(f"{item}\n")
```
